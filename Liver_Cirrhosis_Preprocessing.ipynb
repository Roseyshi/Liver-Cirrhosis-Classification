The prepocessing task will ensure that:




1.   Split the data into test and train data.
2.   Identify if there are null entries and how to handle them.
3.   Identify if there are categorical or numerical data and how to handle the categorical data.
4.   Standardize the data.







```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#
```

```python
data = pd.read_csv('/content/liver_cirrhosis.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('Stage', axis=1), data['Stage'], test_size=0.2, random_state=42)

df_test = pd.concat([X_test, y_test], axis=1)
df_test.to_csv('test_data.csv', index=False)

# Optionally, save the training data if needed
df_train = pd.concat([X_train, y_train], axis=1)
df_train.to_csv('train_data.csv', index=False)
```

```python
train_data = pd.read_csv('/content/train_data.csv')
test_data = pd.read_csv('/content/test_data.csv')
```

```python
y_train.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

X_train_encoded = encoder.fit_transform(train_data[['Status', 'Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']])
X_test_encoded = encoder.transform(test_data[['Status', 'Drug', 'Sex', 'Ascites', 'Hepatomegaly', 'Spiders', 'Edema']])

X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out())
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out())

train_data.reset_index(drop=True, inplace=True)
test_data.reset_index(drop=True, inplace=True)

X_train_encoded = pd.concat([train_data, X_train_encoded_df], axis=1)
X_test_encoded = pd.concat([test_data, X_test_encoded_df], axis=1)

X_train_encoded = X_train_encoded.drop(columns = ['Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema'], axis = 1)
X_test_encoded = X_test_encoded.drop(columns = ['Status','Drug','Sex','Ascites','Hepatomegaly','Spiders','Edema'], axis = 1)

X_train_encoded.head()
```

```python
X_test_encoded.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns = X_train_encoded.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns = X_test_encoded.columns)

X_train_scaled.head()
```
